Hay un conjunto de datos en Kaggle llamado 120 años de historia olímpica. Es un conjunto de datos histórico de los Juegos Olímpicos modernos, desde Atenas 1896 a Rio 2016. Estos datos fueron originalmente recopilados de www.sports-reference.com en mayo del 2018 por un programador R (el código usado para extraer y organizar los data está en GitHub). Nota que los Juegos de Invierno y Verano se llevaron a cabo el mismo año hasta el 1992. Luego de eso, los fueron alternando para que los Juegos de Invierno ocurrieran en un ciclo de cuatro años empezando en el 1994, luego los de Verano en el 1996, después los de Invierno en 1998 y así. Un error común que cometen las personas cuando analizan los datos es asumir que los Juegos de Verano e Invierno siempre han estado alternados.

El conjunto de datos que se facilita aquí es un subconjunto de muestra al azar de ese conjunto de datos que contiene aproximadamente un 15% de las personas que originalmente están en ese conjunto de datos (se le da un ID único a cada persona). Las filas del conjunto de datos a continuación, son de una persona que participó en múltiples eventos en dos juegos olímpicos distintos. 

Tarea
Calcula la estatura promedio de los olimpistas por año. A continuación, un ejemplo de formato de salida. 

```python
#Importo pandas con el siguiente codigo
import pandas as pd
```

```python
#Señalo la ruta del archivo con el siguiente codigo y lo leo en función al tipo de archivo y muestro el mismo
filename = '/content/drive/MyDrive/Data Science 2022/athleteEventsNoPersonal.csv'     # csv
df = pd.read_csv(filename)   # excel
df
```

```python
Estatura_promedio = df.groupby(['ID', 'Year'])[['Height']].mean()
Estatura_promedio
```
