```python
import pandas as pd
import matplotlib.pyplot as plt #from matplotlib import pyplot as plt

filename = '/content/drive/MyDrive/Data Science 2022/excel de pruebas/mortgages.csv'     # csv

df = pd.read_csv(filename)   # csv
df
```

```python
filtro_30 = df['Mortgage Name'] == '30 Year'
filtro_30
```

```python
filtro_003 = df['Interest Rate'] == 0.03
filtro_003
```

```python
filtro_005 = df['Interest Rate'] == 0.05
filtro_005
```

```python
year30003 = (filtro_003) & (filtro_30)
year30003
```

```python
year30005 = (filtro_005) & (filtro_30)
year30005
```

```python
df[year30005]
```

```python
df[year30003]
```

```python
dt = df[year30005][['Interest Paid']].cumsum() 
```

```python
dr = df[year30003][['Interest Paid']].cumsum()
```

```python
df.loc[year30005]['Interest Paid']
```

```python
df['Mortgage Name'].value_counts()
```

```python
from matplotlib import pyplot as plt #import matplotlib.pyplot as plt
#plt.style.use('seaborn')
plt.plot(dt.values, c = 'blue', label = '30 Year 5% APR')
plt.plot(dr.values, c = 'black', label = '30 Year 3% APR')
plt.xticks(rotation = 90);
#plt.ylabel("30 Year")
#plt.xlabel("30 Year")
plt.legend(loc="lower right")
plt.show()
```
